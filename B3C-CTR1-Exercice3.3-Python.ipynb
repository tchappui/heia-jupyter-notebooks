# B3C-CTR1 Exercice 3.3

Tracer le profil de la r√©action bimol√©culaire en phase liquide suivante entre le cyclopentadi√®ne et la benzoquinone pour former triclo‚Äìundec-4,9-di√®ne-3,6-dione:

$$ùê¥ + ùêµ \xrightarrow{k} ùê∂$$

Avec la constante de vitesse k valant 9.92‚àô10-6 m<sup>3</sup>/mol‚àôs, CA0 = 100 mol/m<sup>3</sup> et CB0 = 80 mol/m<sup>3</sup>.


```python
# Importation des packages n√©cessaires
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# D√©finition des constantes
k = 9.92e-6        # s-1
```

```python
# D√©finition des bilans de mati√®res
def bilans(y, t, k):
    CA, CB, CC = y

    # Rate laws
    r = k * CA * CB

    # Rates of transformation of each species
    RA = -r
    RB = -r
    RC = r

    # Partial molar balance equations
    dCA = RA
    dCB = RB
    dCC = RC

    return [dCA, dCB, dCC]
```

```python
# Conditions initiales
CA0 = 100          # mol/m3
CB0 = 80           # mol/m3
CC0 = 0            # mol/m3
```

```python
y0 = [CA0, CB0, CC0]     # [CA0 CB0 CC0]
```

```python
# Domaine d'int√©gration
t = np.linspace(0, 10000, 10001)
```

```python
# R√©solution
y = odeint(bilans, y0, t, (k,))
```

```python
# Visualisation
plt.figure(1, figsize=(8, 6), dpi=100)
plt.plot(t, y)
plt.xlabel('Temps (min)')
plt.ylabel('$C_{i}$ (mol/L)')
plt.legend(['A', 'B', 'C'])
plt.show()
```
