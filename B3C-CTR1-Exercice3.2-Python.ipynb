Une r√©action irr√©versible de premier ordre doit √™tre effectu√©e dans un r√©acteur batch:

$$ùê¥_{1} \overset{k}{\to} ùê¥_{2}+ùê¥_{3}$$

Si ùê∂<sub>ùê¥0</sub>=3 (mol/L) et k = 0.01 (1/s), d√©terminer le profile des concentration de A1 et A2 en fonction du temps et d√©terminer le temps n√©cessaire pour atteindre une conversion de 30 %.


```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Constantes
k = 0.01 # (1/s)
```

```python
# D√©finition des bilans de mati√®re
def bilans(y, t, k):
    C1, C2, C3 = y

    # loi de vitesse
    r = k * C1

    # Vitesses de transformation
    R1 = -r
    R2 = r
    R3 = r

    # Bilans de mati√®re
    dC1 = R1
    dC2 = R2
    dC3 = R3

    return [dC1, dC2, dC3]
```

```python
# Conditions initiales
C10 = 3 # mol/L
C20 = 0 # mol/L
C30 = 0 # mol/L
y0 = [C10, C20, C30] # mol/L
```

```python
# Domaine d'int√©gration
t = np.linspace(0, 300, 301) # s
```

```python
# R√©solution
valeurs = odeint(bilans, y0, t, (k,))
```

```python
# Visualisation
plt.figure(1, figsize=(10, 8), dpi=100)
plt.plot(t, valeurs)
plt.xlabel('$t$ (s)')
plt.ylabel('$C_{i}$ (mol/L)')
plt.legend(['A1', 'A2', 'A3'], loc='best')
plt.show()
```
